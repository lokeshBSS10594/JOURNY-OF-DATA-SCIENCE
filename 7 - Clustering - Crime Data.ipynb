Data Description:

Murder -- Muder rates in different places of United States

UrbanPop - urban population in different places of United States

Rape - Rape rate in different places of United States

# Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#loadign the dataset
crime = pd.read_csv('crime_data.csv')
crime
```

```python
#Normalizing the data
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
#normalized data frame(considering the numerical part of the data)
df_norm=norm_func(crime.iloc[:,1:])
df_norm
```

```python
#creating the dendrogram usning 'scipy.cluster.hierarchy' function
sch.set_link_color_palette(['m','c','y'])
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
#creating the clusetrs
hcluster = AgglomerativeClustering(n_clusters=4,
    affinity='euclidean',
    linkage='complete')
```

```python
hcluster
```

```python
y_hcluster = hcluster.fit_predict(df_norm)
y_hcluster
```

```python
crime ['h_cluster'] = y_hcluster
crime
```

```python
crime[crime['h_cluster']==0]
```

```python
crime[crime['h_cluster']==1]
```

```python
crime[crime['h_cluster']==2]
```

```python
crime[crime['h_cluster']==3]
```

```python
crime.groupby('h_cluster').mean()
```

# Hierarchical clustering - Conclusion

Mudrer's are more in number in the first cluster

Assult's are more in number in the 4th cluster

UrbanPop's are more in number in the 4th cluster and 

Rape's are also more in number in the 4th clusture

THEREFORE, THE CRIME RATE ARE HIGH IN THE 4TH CLUSTER WHEN COMPARED TO OTHER 3 CLUSTERS.

# Kmeans clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime_km = pd.read_csv('crime_data.csv')
crime_km.head(2)
```

```python
import warnings
warnings.filterwarnings('ignore')
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('ELBOW CURVE')
plt.xlabel('CLUSTERS')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime_km['Clust']=x
crime_km
```

```python
crime_km2=crime_km.iloc[:,1:5]
```

```python
crime_km2.groupby(crime_km.Clust).mean()
```

# Kmeans clustering - Conclusion

Mudrer's are more in number in the second cluster

Assult's are more in number in the 4th cluster

UrbanPop's are more in number in the 4th cluster and 

Rape's are also more in number in the 4th clusture

THEREFORE, THE CRIME RATE ARE HIGH IN THE 4TH CLUSTER WHEN COMPARED TO OTHER 3 CLUSTERS.

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime_db=pd.read_csv('crime_data.csv')
crime_db
```

```python
crime_db.info()
```

```python
data_frame=crime_db.iloc[:,1:5]
data_frame
```

```python
data_frame.values
```

```python
stscaler=StandardScaler().fit(data_frame.values)
X=stscaler.transform(data_frame.values)
X
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
```

```python
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cluster=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cluster
```

```python
pd.concat([crime_db,cluster],axis=1)
```
